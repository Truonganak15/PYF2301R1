```python
'''
Bài 13: Cần có tổng 200.000đ từ 3 loại giấy bạc 1000đ, 2000đ, và 5000đ. Lập 
chương tình để tìm tất cả các phương án có thể.
'''
# cach 1
count = 0 # Dem so dap an
total_loop = 0
for i in range (41): #tao ra to hop so luong cac to 5K khac nhau
    for j in range (101): # tao ra to hop so luong cac to 2K khac nhau
        for k in range(201):  # tao ra to hop so luong cac to 1K khac nhau
            total_loop += 1 # co them 1 truong hop kiem tra
            if i*5000+ j*2000 + k*1000 == 200000:
                count += 1 # co them 1 dap an ung
                print (f'Can {i} to 5K, {j} to 2K, {k} to 1k')
print(f'co tong cong {count} cach de doi to 200k')
print (f'co tong cong {total_loop} truong hop kiem tra')
```

```python
'''
Bài 14: Nhập vào một chuỗi s chỉ chứa các ký tự '(', ')', '{', '}', '[' và ']', 
xác định xem chuỗi s có hợp lệ hay không.

Chuỗi đầu vào hợp lệ nếu:

- Các dấu ngoặc mở phải được đóng bằng cùng một loại dấu ngoặc.
- Dấu ngoặc mở phải được đóng theo đúng thứ tự.
- Mọi dấu ngoặc đóng đều có một dấu ngoặc mở tương ứng cùng loại.

Ví dụ 1:
Input: s = "()"
Output: True

Ví dụ 2:
Input: s = "() [] {}"
Output: True

Ví dụ 3:
Input: s = "(]"
Output: False
'''
string = '([{()[{}]}])'
check =''
for c in string:
    # neu c la ky tu mo ngoac
    if c in '([{':
        check += c # luu lai ki tu mo ngoac
    else: # c la ky tu dong ngoac
        if (check[-1] == '(' and c ==')') or (check[-1]=='[' and c == ']') or (check[-1] == '{' and c == '}'):
            check = check[:-1] # bo di ki tu ngoac cuoi cung do da kiem tra xong
        else:
            print(False)
            break
if len(check) == 0:
    print(True)
```
